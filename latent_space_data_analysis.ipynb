```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('latent_space_data_100k')
df.head()
```

```python
df.shape
```

```python
df_labels = df.iloc[:, 512:]
```

```python
# Calculate the counts of each label
counts = df_labels.sum()

# Plot the stacked bar chart
plt.figure(figsize=(8,10))
plt.barh(range(len(counts)), counts, color=['blue', 'orange'], height=0.5)
plt.yticks(range(len(counts)), counts.index)
plt.xlabel('Count')
plt.title('Number of Positives for each Semantic')
plt.show()
```

```python
# Calculate the counts of each label
counts = df_labels.sum()

# Plot the stacked bar chart
plt.figure(figsize=(8,10))
plt.barh(range(len(counts)), counts / len(df_labels), color=['blue', 'orange'], height=0.5)
plt.yticks(range(len(counts)), counts.index)
plt.xlabel('Percentage')
plt.title('Positive Percentage for Semantic in Latent Space Data')
plt.show()
```

```python
df_celeb = pd.read_csv('list_attr_celeba.csv')
df_celeb = df_celeb.replace(-1, 0)
df_celeb = df_celeb.drop('image_id', axis=1)
df_celeb.head()
```

```python
# Calculate the counts of each label
counts = df_celeb.sum()

# Plot the stacked bar chart
plt.figure(figsize=(8,10))
plt.barh(range(len(counts)), counts / len(df_celeb), color=['blue', 'orange'], height=0.5)
plt.yticks(range(len(counts)), counts.index)
plt.xlabel('Percentage')
plt.title('Positive Percentage for Semantic in CelebA')
plt.show()
```
